# Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
x=np.random.rand(50)*100
y=np.random.rand(50)*100
colors=np.random.rand(50)
plt.scatter(x,y,c=colors,cmap="viridis",marker="o")
plt.xlabel("Independent variable")
plt.ylabel("Dependent Variable")
plt.title("Illustration of Scatter Plot")
t=plt.colorbar()
t.set_label("Colorbar",fontsize=15)
```

```python
# Sample Data
advertising_tv = np.array([2, 5, 7, 10, 12, 15, 18, 20, 25, 30, 35, 40])  # TV Ads Spend ($1000)
sales_tv = np.array([5, 9, 11, 14, 18, 24, 27, 30, 35, 38, 45, 50])  # Sales from TV Ads ($1000)

advertising_online = np.array([2, 4, 6, 9, 11, 14, 17, 19, 22, 28, 32, 37])  # Online Ads Spend ($1000)
sales_online = np.array([6, 10, 12, 16, 20, 26, 30, 33, 39, 42, 48, 55])  # Sales from Online Ads ($1000)

plt.scatter(advertising_tv,sales_tv,c="blue",marker="o",label="Advetisment on TV",alpha=0.5)
plt.scatter(advertising_online,sales_online,c="green",marker="s",label="Advetisment Online",alpha=0.5)

plt.xlabel("Advertising Spend ($1000)")
plt.ylabel("Sales Revenue ($1000)")
plt.title("Effect of Advertising Spend on Sales Revenue")

plt.grid(True, linestyle="--", linewidth=0.5, color="gray")
plt.legend()
plt.show()
```
