```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Categories and their sales
categories = ["Electronics", "Clothing", "Groceries", "Furniture", "Toys"]
sales = [5000, 3000, 7000, 2000, 1500]
plt.pie(sales,labels=categories,shadow=True,explode=[0,0,0.1,0,0],autopct="%1.3f%%")
plt.legend(loc=2)
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
categories = ["A", "B", "G", "F", "T"]
sales = [8, 7, 2, 1, 3]
plt.pie(sales,labels=categories,autopct="%1.3f%%",radius=1.5)
plt.pie([1],colors="white")
plt.legend(loc=3)
plt.show()
```
