# Bar Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data.csv',usecols=["Product","Sales"])
```

```python
df=pd.DataFrame(data)
df.columns
```

```python
plt.bar(df["Product"],df["Sales"],color="green",label="Sales",width=.5)
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales Data")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", linewidth=0.5, color="gray")
plt.legend(labelcolor="red",edgecolor="yellow")#facecolor="blue"
plt.show()
```

```python
categories = ["fan", "cooler", "Ac", "Window Ac"]
values1 = [20, 35, 30, 50] 
values2 = [25, 32, 34, 40]
arr=np.arange(len(categories))
w=0.4
plt.bar(arr-w/2,values1,color="skyblue",label="values1",width=w)
plt.bar(arr+w/2,values2,color="orange",label="values2",width=w)

plt.xlabel("Electrical appliace")
plt.ylabel("Sales")
plt.xticks(arr,categories,rotation=45)

plt.legend()
plt.show()
```

```python
categories = ["fan", "cooler", "Ac", "Window Ac"]
values1 = [20, 35, 30, 50] 
values2 = [25, 32, 34, 40]
w=0.3

y = np.arange(len(categories)) 

plt.barh(y-w/2,values1,color="red",label="values1",height=w)
plt.barh(y+w/2,values2,color="black",label="values2",height=w)

plt.title("Horizontal bar graph")
plt.ylabel("Electrical appliace")
plt.xlabel("Sales")

plt.grid(True, linestyle="--", linewidth=0.5, color="gray")
plt.yticks(y,categories)
plt.legend()
plt.show()
```
