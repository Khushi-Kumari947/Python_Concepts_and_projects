# Box plot and Whisker Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = [20, 21, 19, 18, 30, 45, 23, 25, 22, 22, 28, 32]

# Create box plot
plt.boxplot(data,vert=False)
plt.title("Box and Whisker Plot")
plt.ylabel("Values")
plt.show()
```

```python
# Sample data
data = [20, 21, 19, 18, 30, 45, 23, 25, 22, 22, 28, 32]

# Create box plot
sns.boxplot(data=data,orient="h")
plt.title("Box Plot using Seaborn")
plt.xlabel("Values")
plt.show()
```

```python
# Sample DataFrame
df = pd.DataFrame({
    'Math': [70, 75, 80, 65, 90, 85, 78, 92, 60, 88],
    'Science': [72, 78, 83, 69, 91, 86, 80, 95, 63, 89]
})

# Create box plot
df.boxplot(column=['Math', 'Science'])
plt.title("Subject-wise Box Plot")
plt.ylabel("Scores")
plt.show()
```

```python
# Sample DataFrame
data = {
    'Subject': ['Math'] * 10 + ['Science'] * 10 + ['English'] * 10,
    'Score': [78, 85, 82, 90, 76, 88, 84, 79, 91, 77,
              70, 75, 72, 80, 74, 68, 77, 73, 69, 81,
              88, 90, 85, 92, 87, 84, 91, 89, 86, 93]
}
df = pd.DataFrame(data)
df
```

```python
sns.set(style="whitegrid")

# Create box plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Subject', y='Score',data=df, palette='pastel')

# Add Matplotlib customizations
plt.title("Box Plot of Student Scores by Subject", fontsize=14)
plt.xlabel("Subject", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.ylim(60, 100)  # Set y-axis range

# Show plot
plt.tight_layout()
plt.show()
```

```python
Maths=[70, 75, 80, 65, 90, 85, 78, 92, 60, 88]
Science=[72, 78, 83, 69, 91, 86, 80, 95, 63, 89]
y=[Maths,Science]
plt.boxplot(y,tick_labels=["Maths","Science"],showmeans=True,patch_artist=True)
plt.show()
```
